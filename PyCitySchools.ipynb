```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
school_data_complete.rename(
    columns={
        'student_name':'Student Name', 'gender':'Gender', 'grade':'Grade', 
        'school_name':'School Name', 'reading_score':'Reading Score', 'math_score':'Math Score', 
        'type':'Type', 'size':'Size', 'budget':'Budget'},
    inplace=True)
school_data_complete.head()
```

```python
# Reorganize columns
school_data_complete = school_data_complete[['Student Name', 'Student ID', 'Grade', 
                                             'Gender', 'Reading Score', 'Math Score', 
                                             'School Name', 'School ID', 'Type', 'Size', 'Budget', 'Grade']]
school_data_complete.set_index('Student Name', inplace=True)
school_data_complete.head()
```

```python
school_data_complete.drop(['School ID'], axis=1, inplace=True)
```

```python
# Calculate the total number of schools 
schools_count = school_data_complete['School Name'].nunique()
schools_count
```

```python
# List all schools
schools_list = school_data_complete['School Name'].unique()
schools_list
```

```python
# Calculate the total number of students
student_count = school_data_complete.shape[0]
student_count
```

```python
# Calculate total budget
total_budget = f'${school_data_complete["Budget"].unique().sum()}'
total_budget
```

```python
# Calculate average math score
average_math_score = school_data_complete['Math Score'].mean()
average_math_score
```

```python
# Calculate average reading score
average_reading_score = round(school_data_complete['Reading Score']
                              .mean(), 6)
average_reading_score
```

```python
average_overall_score = round((average_reading_score + 
                               average_math_score)/2, 6)
average_overall_score
```

```python
passing_math_students = school_data_complete.loc[(school_data_complete[
    'Math Score'] >= 70)]
passing_math_students.head()                             
```

```python
# Calculate percentage of students with passing math score
passing_math_students_percentage = round(((len(passing_math_students)/
                                        len(school_data_complete[
                                        'Math Score']))) * 100, 6)
passing_math_students_percentage
```

```python
passing_reading_students = school_data_complete.loc[(school_data_complete["Reading Score"] >= 70)]
passing_reading_students.head()
```

```python
# Calculate percentage of students with passing reading score
passing_reading_students_percentage = round((len(passing_reading_students)/
                                             len(school_data_complete['Reading Score'])) * 100, 6)
passing_reading_students_percentage
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = round((passing_math_students_percentage + passing_reading_students_percentage)/2, 6)
overall_passing_rate
```

```python
# Create a summary dataframe
# The directions indicate "overall passing rate" should be calculated as the the overall average score, 
# which didn't make sense to me.
# I created two different variables "average overall score" and "overall passing rate" which shows there is a 
# small difference between those two variables. 

summary_df = pd.DataFrame({
    'Total Schools':[schools_count],
    'Total Students':[student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_students_percentage],
    '% Passing Reading': [passing_reading_students_percentage],
    '% Overall Passing Rate': [overall_passing_rate],
    'Average Overall Score': [average_overall_score]
})
summary_df
```

```python
school_data_complete['Student Passed Math'] = np.where(school_data_complete[
    'Math Score']>=70, True, False)
school_data_complete.head()
```

```python
school_data_complete['Student Passed Reading'] = np.where(school_data_complete[
    'Reading Score']>=70, True, False)
school_data_complete.head()
```

```python
school_data_complete['Type = Charter'] = np.where(school_data_complete['Type']=='Charter', True, False)
school_data_complete.head()
```

```python
# Grab school data columns
school_data = school_data_complete[['School Name', 'Type', 'Budget', 'Math Score', 
                                    'Reading Score', 'Student ID', 'Student Passed Math', 
                                    'Student Passed Reading', 'Type = Charter']]
school_data['Student ID'] = 1
school_data.head()
```

```python
# Group by school/find average scores
school_groups = school_data.groupby(['School Name'], sort=False).agg(
    {'Math Score':np.mean,
    'Reading Score':np.mean,
    'Student ID':np.sum,
    'Student Passed Math':np.sum,
    'Student Passed Reading':np.sum,
    #'Type = Charter':np.
    })
school_groups.head()
```

```python
school_groups['% Passing Math'] = (school_groups['Student Passed Math']/school_groups['Student ID']) * 100
school_groups.head()
```

```python
school_groups['% Passing Reading'] = (school_groups['Student Passed Reading']/school_groups['Student ID']) * 100
school_groups.head()
```

```python
school_groups['% Overall Passing Rate'] = (school_groups['% Passing Math'] + 
                                         school_groups['% Passing Reading'])/2
school_groups.head()
```

```python
school_groups['Total School Budget'] = school_data_complete["Budget"].unique()
school_groups.head()
```

```python
school_groups = school_groups.rename(columns=
                                    {'Student ID':'Total Students',
                                    'Math Score':'Average Math Score',
                                    'Reading Score':'Average Reading Score'
                                    })
school_groups.head()
```

```python
school_groups['Per Student Budget'] = (school_groups['Total School Budget']/school_groups['Total Students'])
school_groups['Per Student Budget'] = list(map(int, school_groups['Per Student Budget']))
school_groups
```
